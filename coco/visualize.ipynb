```python
%matplotlib inline
from pycocotools.coco import COCO
import numpy as np
import matplotlib.pyplot as plt
import pylab
import skimage.io as io
import os
pylab.rcParams['figure.figsize'] = (8.0, 10.0)
```

```python
dataDir='/home/sung/dataset/assembly_group'
dataType='val2017'
annFile='{}/annotations/group_{}.json'.format(dataDir,dataType)
```

```python
# initialize COCO api for instance annotations
coco=COCO(annFile)
```

```python
cat = coco.loadCats(coco.getCatIds())
print(cat)
```

```python
# get all images containing given categories, select one at random
catIds = coco.getCatIds(catNms=['assembly']);
imgIds = coco.getImgIds(catIds=catIds);
print(imgIds)
img = coco.loadImgs(imgIds[0])[0]
print(img)
```

```python
I = io.imread(os.path.join('train2017', img['file_name']))
plt.axis('off')
plt.imshow(I)
plt.show()
```

```python
# load and display instance annotations
plt.imshow(I); plt.axis('off')
annIds = coco.getAnnIds(imgIds=img['id'], catIds=catIds, iscrowd=None)
anns = coco.loadAnns(annIds)
print(anns)
coco.showAnns(anns, draw_bbox=True)
```
